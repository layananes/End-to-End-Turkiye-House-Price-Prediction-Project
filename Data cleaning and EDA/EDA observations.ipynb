I will finally create a correlation matrix and observe some last columns

```python
from dython.nominal import associations
complete_correlation = associations(df, figsize=(12, 12))
```

As expected, most correlation is seen with net area, gross area, room count, and bathroom count. Though what is most surprising is the correlation with building age, heating type and floor location. With the plot shown next however, I find the correlation with floor location a little unusual. 

```python
correlation_series = (
    df.groupby('Floor_Location')[['Price', 'Net_Area']]
      .corr()
      .unstack()
      .iloc[:, 1]
      .sort_values(ascending=False)
)

print("\nPrice vs. Net_Area Pearson Correlation by Building Age")
print(correlation_series.to_string(float_format='%.4f'))
```

```python
plt.figure(figsize=(38, 10))
df.plot.scatter(x='Floor_Location', y='Price',
    title='Pandas Scatter Plot Example',
    grid=True,
    ax=plt.gca()
)
plt.show()
```

In the last section, I attempted to observe the correlation of net are and price in every category of heating type and building age. It has, to a degree, solidified my observations, since a relatively high correlation between net area and price can be seen across the "good" categories. 

For example, floor heating has pearson correlation of almost 0.80, which is very high compared to worse heating types like soba or natural gas. Building ages that are newer are also almost all around 0.7, which is higher than the correlation in older buildings (which are lower at around 0.5-0.6)
