**handling the Floor Location**

so for the missing floor location i am thinking of calculating the most frequent for each catogery, what i mean by this is if the bulding is 2 floors the location can not exceed the 2nd floor so we calculate the most frequent location for the these 2floors' bulding and put it in these NaN rows and so on for the rest, that is what i am going to do 

```python
# 1. Create a "Lookup Table"
# Group by 'Total_Building_Floors' and find the most common 'Floor_Location'
floor_modes = data.groupby('Total_Building_Floors')['Floor_Location'].apply(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
)

# 2. Fill the missing values
# Map the Total Floors to the mode we found
data['Floor_Location'] = data['Floor_Location'].fillna(data['Total_Building_Floors'].map(floor_modes))
```

**Handling the missing rooms**

so now i want to handle the NaN in the room counts the same way that we did with the floor location, but instead using the gross area as a referance

```python
# 1. Define Size Bins (Ranges)
# We group areas into logical categories: 0-50m, 50-75m, etc.
bins = [-1, 50, 75, 100, 125, 150, 200, float('inf')]
labels = [1, 2, 3, 4, 5, 6, 7] # Labels for the groups

# 2. Create a temporary helper column
data['Area_Group'] = pd.cut(data['Gross_Area'], bins=bins, labels=labels)

# 3. Calculate the Mode (Most Common Room Count) for each size group
room_modes = data.groupby('Area_Group', observed=True)['Room_Count'].apply(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
)

print("Most Common Rooms per Size:")
print(room_modes)

# 4. Fill the Missing Values
# Look up the correct room count based on the area group
data['Room_Count'] = data['Room_Count'].fillna(data['Area_Group'].map(room_modes))

# 5. Clean up (Remove the helper column)
data.drop(columns=['Area_Group'], inplace=True)

# Final Check
print(f"Remaining Missing Room_Count: {data['Room_Count'].isnull().sum()}")
```

#### **_________________________________________________________________________________________________**

**Handling Bathroom_Count**

    i will handle it the same way i handled the other (by finding the common count in each group and replace it with NaN)

```python
data_copy = data.copy()
```

```python
# 1. Define a function to categorize the rooms
def categorize_rooms(rooms):
    if rooms <= 3:
        return "1-3 Rooms"
    elif rooms == 4:
        return "4 Rooms"
    else:
        return "5+ Rooms"

# 2. Apply it to create a temporary helper column
data['Room_Group'] = data['Room_Count'].apply(categorize_rooms)

# 3. Create a Crosstab with Percentages
# normalize='index' turns counts into percentages (0.87 instead of 5354)
analysis = pd.crosstab(data['Room_Group'], data['Bathroom_Count'], normalize='index') * 100

# Show only 1, 2, and 3 bathrooms for clarity
print(analysis[[1.0, 2.0, 3.0]].round(1))
```

```python
# Create the full frequency table
cross_tab = pd.crosstab(data['Room_Count'], data['Bathroom_Count'])

# Filter only for 3, 4, and 5 rooms to see the transition
print(cross_tab.loc[[3.0, 4.0, 5.0]])
```

as we see here there is a relation between the room_counts and the bathroom_counts

```python
# 1. Create the Lookup Table (Mode)
# Group by Room Count and find the most common Bathroom Count
bathroom_modes = data.groupby('Room_Count')['Bathroom_Count'].apply(
    lambda x: x.mode().iloc[0] if not x.mode().empty else None
)

print("Mapping (Room Count -> Most Common Bathrooms):")
print(bathroom_modes)

# 2. Fill the Missing Values
data['Bathroom_Count'] = data['Bathroom_Count'].fillna(data['Room_Count'].map(bathroom_modes))

# 3. Final Safety Net (for any weird room counts not in the map)
# Fill remaining with 1.0 (the global mode)
data['Bathroom_Count'] = data['Bathroom_Count'].fillna(1.0)

# Final Check
print(f"Remaining Missing Bathrooms: {data['Bathroom_Count'].isnull().sum()}")
```

**Final notes on the NaN handling**

- i used a the "Conditional Mode" strategy (grouping by X to fill Y) 
- the main idea is finde the most used data in specific group the replace the NaN according to it
- the data that we used to predict the most frequent X in some columns should be mostly complete 

- we can use "Machine Learning Imputer" (like KNN), but this will be overkill, and unnecessary work since it wont affect the model

```python
def check_outliers(df, col_name, target_name='Price'):
    """
    Creates side-by-side Box Plot and Scatter Plot to detect outliers.
    
    Parameters:
    - df: Your dataframe (e.g., data)
    - col_name: The column to check for outliers (e.g., 'Net_Area')
    - target_name: The target variable for scatter plot (default: 'Price')
    """
    plt.figure(figsize=(16, 6))

    # Plot 1: Box Plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df[col_name], color='skyblue', linewidth=2)
    plt.title(f'Box Plot of {col_name}', fontsize=14, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)

    # Plot 2: Scatter Plot
    plt.subplot(1, 2, 2)
    sns.scatterplot(x=df[col_name], y=df[target_name], color='coral', alpha=0.6, s=50)
    plt.title(f'{col_name} vs {target_name}', fontsize=14, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Optional: Log scale if Price is the target (helps readability)
    # plt.yscale('log') 

    plt.tight_layout()
    plt.show()
```
